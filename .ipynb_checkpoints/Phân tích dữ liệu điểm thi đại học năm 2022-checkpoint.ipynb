```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import seaborn as sns
%matplotlib inline
```

Check qua dữ liệu điểm thi năm 2022 và thực hiện data wrang|ling nhé

```python
df_2022 = pd.read_csv("diem_thi_thpt_2022.csv")
df_2022.shape
```

```python
df_2022.head()
```

```python
#Function lấy city code từ số báo danh
def cityCode(sbd):
    return int(str(sbd)[:-6])
```

```python
#thêm cột city_code để phân tích dữ liệu theo tỉnh thành
df_2022['city_code'] = df_2022["sbd"].apply(cityCode)
```

```python
#Lấy dữ liệu city name theo city code
df_city = pd.read_csv('city_code.csv')
df_city
```

```python
# Thêm City Name vào cột dữ liệu năm 2022 dựa vào city code
df_2022 = df_2022.merge(df_city, how='left', on='city_code')
```

```python
df_2022.head()
```

```python
df_2022.tail()
```

```python
df_2022.describe()
```

Check dữ liệu điểm thi năm 2021, format tương tự với 2022

```python
df_2021 = pd.read_csv("diem_thi_thpt_2021.csv")
```

```python
df_2021.head()
```

```python
#drop dữ liệu không sử dụng đến
df_2021.drop('Unnamed: 0', axis=1, inplace=True)

#Đổi tên các column sang dạng lowercase để đồng bộ với dữ liệu năm 2022
df_2021.columns = list(map(lambda x: x.lower(), df_2021.columns.to_list()))

#Đổi cột cum_thi thành city_code
df_2021.rename(columns={"cum_thi":"city_code"}, inplace=True)
```

```python
df_2021 = df_2021.merge(df_city, how='left', on='city_code')
```

```python
df_2021.head()
```

```python
df_2021.tail()
```

```python
df_2021.describe()
```

Đầu tiên chúng ta sẽ phân tích thử môn toán theo phổ điểm bằng biểu đồ density

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax = sns.kdeplot(data=df_2022, x='toan', color='red', label='2022')

ax = sns.kdeplot(data=df_2021, x='toan', color='blue', label='2021')

ax.set_title('Biểu đồ density của đề thi toán qua các năm')
ax.legend()
plt.show()
```

```python
print('mode năm 2021: ', df_2021['toan'].mode())

print('mode năm 2022: ', df_2022['toan'].mode())
```

Nhìn chung, điểm môn toán trong cả hai năm không có nhiều khác biệt, có chăng thì số câu khó (câu lấy điểm 8, 9, 10) của năm 2022 sẽ nhiều hơn số câu khó của năm 2021 một chút. Mode (số điểm xuất hiện nhiều nhất) của cả 2 năm đều rơi vào khoảng 7.8. Có thể thấy đỉnh của biểu đồ khá dốc => đề thi hầu như chỉ phân loại ra các câu dễ-trung bình và câu khó, không có nhiều câu để phân loại học sinh khá => khả năng phân loại của đề thi toán chưa tốt

Chúng ta thử sang phổ điểm của đề thi văn nhé

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax = sns.kdeplot(data=df_2022, x='ngu_van', color='red', label='2022', bw_method=0.13)

ax = sns.kdeplot(data=df_2021, x='ngu_van', color='blue', label='2021', bw_method=0.13)

ax.set_title('Biểu đồ density của đề thi Văn qua các năm')
ax.legend()
plt.show()
```

```python
print('mode năm 2021: ', df_2021['ngu_van'].mode())

print('mode năm 2022: ', df_2022['ngu_van'].mode())
```

Nhìn chung thì điểm Văn của hai năm không có quá nhiều khác biệt, có vẻ như đề thi năm 2022 có tính phân loại tốt hơn khi đỉnh không cao bằng năm 2021, nhưng đỉnh vẫn còn khá dốc.

Sang thử đề thi tiếng anh nhé

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax = sns.kdeplot(data=df_2022, x='ngoai_ngu', color='red', label='2022')

ax = sns.kdeplot(data=df_2021, x='ngoai_ngu', color='blue', label='2021')

ax.set_title('Biểu đồ density của đề thi Tiếng Anh qua các năm')
ax.legend()
plt.show()
```

```python
print('mode năm 2021: ', float(df_2021['ngoai_ngu'].mode()))
print('Đỉnh thứ 2 năm 2021: ', float(df_2021[df_2021['ngoai_ngu'] >  6]['ngoai_ngu'].mode()))
print('mean năm 2021: ', df_2021['ngoai_ngu'].mean())
print()

print('mode năm 2022: ', float(df_2022['ngoai_ngu'].mode()))
print('mean năm 2022: ', df_2022['ngoai_ngu'].mean())
```

Có thể thấy điểm tiếng anh năm 2021 có tới 2 đỉnh (4.0 và 9.0), đây là hai đỉnh đại diện cho hai nhóm là nhóm học sinh yếu và nhóm học sinh giỏi, nhóm sinh giỏi tập trung nhiều tại các tỉnh, thành phố lớn nơi có chất lượng dạy và học tốt hơn, đồng thời sử dụng tiếng anh nhiều hơn trong công việc, ngược lại nhóm học sinh yếu tập trung chủ yếu ở các tỉnh, vùng nông thôn chưa chú trọng đầu tư nhiều vào môn tiếng anh. Sang đến năm 2022, điểm tiếng anh chỉ còn lại một đỉnh nhưng mode (3.8) và mean (5.16) lại nhỏ hơn so với năm 2021, điều này cho thấy đề thi năm 2022 khó hơn tương đối chủ yếu do các câu hỏi lấy điểm cao (8, 9, 10) khó hơn, và do đó đề thi cũng có tính phân loại tốt hơn năm 2021.

Bây giờ thử plot một vài Choropleth map để xem có tìm được insignt nào không nhé, chúng ta sẽ bắt đầu với môn toán

```python
#import dữ liệu địa lý (long, lat), mình sẽ lấy dữ liệu của Open Development Vietnam và sử dụng thư viện geopandas để xử lý
geo_data = gpd.read_file(r"C:\Users\User\Desktop\Data analysis course\Geographic\vietnam_shape\gadm36_VNM_1.shp")
geo_data
```

```python
df_toan = df_2022.groupby('city_name')['toan'].agg(['mean', 'count', 'std']).reset_index()
```

```python
#Tạo một dataframe đếm số lượng hsg toán ở các tỉnh
df_count_if = df_2022[df_2022['toan'] >= 8].groupby('city_name').count()['toan'].to_frame().reset_index().rename(columns={'toan':'hsg'})

#merge dữ liệu với df_toan và thêm cột tỷ lệ
df_toan = pd.merge(df_toan,df_count_if, how='left', on='city_name')
df_toan['ty_le_hsg'] = df_toan['hsg']/df_toan['count']*100
```

```python
#Bây giờ, chúng ta sẽ merge dữ liệu của geo_data với dữ liệu điểm trung bình, phương sai, tỷ lệ điểm trên 8 của môn toán theo cột name
geo_toan =  geo_data.merge(df_toan, how='left', left_on='city_name', right_on='city_name')
geo_toan
```

```python
fig, axis = plt.subplots(1,3,figsize=(20, 15))
fig.suptitle("Môn Toán", x=0.5, y=0.87, fontsize='xx-large')

geo_toan.plot('mean', cmap="Blues", legend = False, ax=axis[0])
axis[0].set_title('Điểm trung bình theo vùng miền')
axis[0].axis('off')

geo_toan.plot('ty_le_hsg', cmap="Greens", legend = False, ax=axis[1])
axis[1].set_title('Tỷ lệ học sinh giỏi theo vùng miền')
axis[1].axis('off')

geo_toan.plot('std', cmap="Reds", legend = False, ax=axis[2])
axis[2].set_title('Phương sai theo vùng miền')
axis[2].axis('off')

plt.show()
```

Tương tự với môn văn nhé

```python
#Tạo một dataframe cho môn văn để group dữ liệu
df_ngu_van = df_2022.groupby('city_name')['ngu_van'].agg(['mean', 'count', 'std']).reset_index()

#Tạo một dataframe đếm số lượng hsg văn ở các tỉnh
df_count_if = df_2022[df_2022['ngu_van'] >= 8].groupby('city_name').count()['ngu_van'].to_frame().reset_index().rename(columns={'ngu_van':'hsg'})

#merge dữ liệu với df_ngu_van và thêm cột tỷ lệ
df_ngu_van = pd.merge(df_ngu_van,df_count_if, how='left', on='city_name')
df_ngu_van['ty_le_hsg'] = df_ngu_van['hsg']/df_ngu_van['count']*100

#Bây giờ, chúng ta sẽ merge dữ liệu của geo_data với dữ liệu điểm trung bình, phương sai, tỷ lệ điểm trên 8 của môn văn theo cột name
geo_ngu_van =  geo_data.merge(df_ngu_van, how='left', left_on='city_name', right_on='city_name')
geo_ngu_van
```

```python
fig, axis = plt.subplots(1,3,figsize=(20, 15))
fig.suptitle("Môn Ngữ Văn", x=0.5, y=0.87, fontsize='xx-large')

geo_ngu_van.plot('mean', cmap="Blues", legend = False, ax=axis[0])
axis[0].set_title('Điểm trung bình theo vùng miền')
axis[0].axis('off')

geo_ngu_van.plot('ty_le_hsg', cmap="Greens", legend = False, ax=axis[1])
axis[1].set_title('Tỷ lệ học sinh giỏi theo vùng miền')
axis[1].axis('off')

geo_ngu_van.plot('std', cmap="Reds", legend = False, ax=axis[2])
axis[2].set_title('Phương sai theo vùng miền')
axis[2].axis('off')

plt.show()
```

Môn Tiếng Anh

```python
#Tạo một dataframe cho môn văn để group dữ liệu
df_ngoai_ngu = df_2022.groupby('city_name')['ngoai_ngu'].agg(['mean', 'count', 'std']).reset_index()

#Tạo một dataframe đếm số lượng hsg văn ở các tỉnh
df_count_if = df_2022[df_2022['ngoai_ngu'] >= 8].groupby('city_name').count()['ngoai_ngu'].to_frame().reset_index().rename(columns={'ngoai_ngu':'hsg'})

#merge dữ liệu với df_ngu_van và thêm cột tỷ lệ
df_ngoai_ngu = pd.merge(df_ngoai_ngu,df_count_if, how='left', on='city_name')
df_ngoai_ngu['ty_le_hsg'] = df_ngoai_ngu['hsg']/df_ngoai_ngu['count']*100

#Bây giờ, chúng ta sẽ merge dữ liệu của geo_data với dữ liệu điểm trung bình, phương sai, tỷ lệ điểm trên 8 của môn văn theo cột name
geo_ngoai_ngu =  geo_data.merge(df_ngoai_ngu, how='left', left_on='city_name', right_on='city_name')
geo_ngoai_ngu
```

```python
fig, axis = plt.subplots(1,3,figsize=(20, 15))
fig.suptitle("Môn Tiếng Anh", x=0.5, y=0.87, fontsize='xx-large')

geo_ngoai_ngu.plot('mean', cmap="Blues", legend = False, ax=axis[0])
axis[0].set_title('Điểm trung bình theo vùng miền')
axis[0].axis('off')

geo_ngoai_ngu.plot('ty_le_hsg', cmap="Greens", legend = False, ax=axis[1])
axis[1].set_title('Tỷ lệ học sinh giỏi theo vùng miền')
axis[1].axis('off')
geo_ngoai_ngu.plot('std', cmap="Reds", legend = False, ax=axis[2])
axis[2].set_title('Phương sai theo vùng miền')
axis[2].axis('off')

plt.show()
```
